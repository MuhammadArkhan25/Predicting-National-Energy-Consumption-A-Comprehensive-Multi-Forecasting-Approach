##Penjelasan Umum##

Key Features:



*   **Entity:** Nama negara atau wilayah tempat data dilaporkan.
*   **Year**: Tahun dimana data dilaporkan, berkisar antara tahun 2000 hingga 2020.
*   **Access to electricity (% of population)**: Persentase penduduk yang memiliki akses terhadap listrik.
*   **Access to clean fuels for cooking (% of population):** Persentase penduduk yang sangat bergantung pada bahan bakar ramah lingkungan.
* **Renewable-electricity-generating-capacity-per-capita:** Kapasitas energi terbarukan terpasang per orang
*   **Financial flows to developing countries (US$):** Aid and assistance Bantuan dan bantuan dari negara maju untuk proyek energi bersih.

*   **Electricity from fossil fuels (TWh):** Listrik dihasilkan dari bahan bakar fosil (batubara, minyak, gas) dalam terawatt-jam.
*   **Electricity from nuclear (TWh):** Listrik yang dihasilkan dari tenaga nuklir dalam terawatt-jam.

*   **Electricity from renewables (TWh):** Listrik dihasilkan dari sumber terbarukan (hidro, surya, angin, dll.) dalam terawatt-jam.
*   **Low-carbon electricity (% electricity):** Persentase listrik dari sumber rendah karbon (nuklir dan energi terbarukan).
*  **Primary energy consumption per capita (kWh/person):** Konsumsi energi per orang dalam kilowatt-jam.
*   **Energy intensity level of primary energy (MJ/$2011 PPP GDP):** Penggunaan energi per unit PDB pada paritas daya beli.
*   **Value_co2_emissions (metric tons per capita):** Emisi karbon dioksida per orang dalam metrik ton
*   **Renewables (% equivalent primary energy):** Energi primer setara yang berasal dari sumber terbarukan.
*   **GDP growth (annual %):** Tingkat pertumbuhan PDB tahunan berdasarkan mata uang lokal yang konstan.
*   **GDP per capita:** Produk domestik bruto per orang.
*  **Density (P/Km2):** Kepadatan penduduk dalam satuan orang per kilometer persegi.
*  **Land Area (Km2):** Total luas daratan dalam kilometer persegi.
*  **Latitude:** Latitude of the country's centroid in decimal degrees.
*  **Longitude:** Longitude of the country's centroid in decimal degrees.

```python
# Impor library Darts
import pandas as pd
from google.colab import drive
import seaborn as sns
import matplotlib.pyplot as plt
import geopandas as gpd
import plotly_express as px
from darts import TimeSeries
from darts.models import NBEATSModel
from darts.dataprocessing.transformers import Scaler
from darts.metrics import mape
from darts.utils.timeseries_generation import datetime_attribute_timeseries
import numpy as np
```

```python
# Pilih negara-negara yang akan dimodelkan
countries = ['South Africa', 'China', 'India', 'Brazil','Indonesia']
```

```python
# Buat dictionary untuk menyimpan TimeSeries untuk setiap negara
series_dict = {}
```

```python
# Loop melalui setiap negara dan buat TimeSeries
for country in countries:
    df_country = df[df['Entity'] == country]
    series = TimeSeries.from_dataframe(df_country, 'Year', 'Primary energy consumption per capita (kWh/person)')
    series_dict[country] = series
```

```python
# Normalisasi Data
transformer = Scaler()
series_dict_scaled = {country: transformer.fit_transform(series) for country, series in series_dict.items()}
```

```python
# Pemodelan
# Inisialisasi model N-BEATS
model = NBEATSModel(
    input_chunk_length=5,  # Panjang input sequence dikurangi
    output_chunk_length=2,  # Panjang output sequence dikurangi
    random_state=42  # Untuk reproduksibilitas
)
```

```python
# Prediksi
# Tentukan horizon prediksi (misalnya, 5 tahun ke depan)
horizon = 5

# Latih model untuk setiap negara
for country, series in series_dict_scaled.items():
    # Pastikan data pelatihan cukup panjang untuk model
    if len(series) > model.input_chunk_length:
        model.fit(series[:-horizon])  # Jangan gunakan data horizon untuk pelatihan
    else:
        print(f"Skipping {country} due to insufficient data for training.")
```

```python
# Buat dictionary untuk menyimpan prediksi untuk setiap negara
predictions_dict = {}
```

```python
# Loop melalui setiap negara dan buat prediksi
for country, series in series_dict_scaled.items():
    # Pastikan data cukup panjang untuk prediksi
    if len(series) > model.input_chunk_length:
        predictions = model.predict(n=horizon, series=series[:-horizon])  # Gunakan data hingga sebelum horizon
        predictions_dict[country] = transformer.inverse_transform(predictions)
    else:
        print(f"Skipping prediction for {country} due to insufficient data.")
```

```python
# Evaluasi
model = NBEATSModel(
    input_chunk_length=10,  # Menambah panjang input sequence dapat meningkatkan akurasi
    output_chunk_length=3,  # Menyesuaikan panjang output sequence
    num_stacks=10,  # Menambah jumlah stacks dapat meningkatkan kapasitas model
    num_blocks=3,  # Menyesuaikan jumlah blocks per stack
    num_layers=4,  # Menambah jumlah layers per block
    layer_widths=512,  # Menambah lebar layer
    random_state=42  # Untuk reproduksibilitas
)

# Hitung MAPE (Mean Absolute Percentage Error) untuk setiap negara
for country, prediction in predictions_dict.items():
    historical_series = series_dict[country]
    # Pastikan ada data historis yang cukup untuk dibandingkan dengan prediksi
    if len(historical_series) > horizon:
        mape_score = mape(historical_series[-horizon:], prediction)
        print(f"MAPE for {country}: {mape_score:.2f}%")
    else:
        print(f"Skipping MAPE calculation for {country} due to insufficient historical data.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Visualisasi
plt.figure(figsize=(12, 6))  # Adjust figure size for better visualization

for country, prediction in predictions_dict.items():
    historical_series = series_dict[country]

    # Dapatkan indeks waktu terakhir dari data historis
    last_historical_time = historical_series.time_index[-1]

    # Buat indeks waktu untuk prediksi, mulai dari setelah data historis berakhir
    prediction_time_index = pd.date_range(start=last_historical_time + pd.DateOffset(years=1), periods=horizon, freq='Y')

    # Gabungkan data historis dan prediksi untuk plotting yang berkelanjutan
    combined_time_index = historical_series.time_index.append(prediction_time_index)
    combined_values = np.concatenate((historical_series.values().flatten(), prediction.values().flatten()))

    # Plot data gabungan
    plt.plot(combined_time_index, combined_values, label=country)

    # Plot prediksi dengan warna yang sama tetapi garis putus-putus, mulai dari titik terakhir data historis
    plt.plot(prediction_time_index, prediction.values().flatten(), linestyle='--', color=plt.gca().lines[-1].get_color(), label=f"{country} Prediction")

plt.title('Konsumsi Energi Primer per Kapita')
plt.xlabel('Tahun')
plt.ylabel('kWh/person')
plt.legend()
plt.grid(True)
plt.show()

```

```python
# Buat dictionary untuk menyimpan model Prophet untuk setiap negara
prophet_models = {}
```

```python
# Loop melalui setiap negara dan buat model Prophet
for country in countries:
    df_country = df[df['Entity'] == country].copy()
    df_country.rename(columns={'Year': 'ds', 'Primary energy consumption per capita (kWh/person)': 'y'}, inplace=True)
    model = Prophet()
    model.fit(df_country)
    prophet_models[country] = model
```

```python
# Prediksi
# Tentukan horizon prediksi (misalnya, 5 tahun ke depan)
horizon = 5
future_dates = prophet_models[countries[0]].make_future_dataframe(periods=horizon, freq='Y')
```

```python
# Buat dictionary untuk menyimpan prediksi Prophet untuk setiap negara
prophet_predictions = {}
```

```python
# Loop melalui setiap negara dan buat prediksi
for country, model in prophet_models.items():
    forecast = model.predict(future_dates)
    prophet_predictions[country] = forecast[['ds', 'yhat']][-horizon:]
```

```python

# Visualisasi
plt.figure(figsize=(12, 6))

for country, prediction in prophet_predictions.items():
    historical_series = series_dict[country]

    # Plot data historis
    plt.plot(historical_series.time_index, historical_series.values().flatten(), label=country)

    # Plot prediksi Prophet
    plt.plot(prediction['ds'], prediction['yhat'], linestyle='--', color=plt.gca().lines[-1].get_color(), label=f"{country} Prophet Prediction")

plt.title('Konsumsi Energi Primer per Kapita (Prophet)')
plt.xlabel('Tahun')
plt.ylabel('kWh/person')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Hitung MAPE untuk setiap negara (Prophet)
for country, prediction in prophet_predictions.items():
    historical_series = series_dict[country]
    actual_values = historical_series.values().flatten()[-horizon:]  # Ambil nilai aktual untuk horizon prediksi
    predicted_values = prediction['yhat'].values

    # Hitung MAPE hanya jika ada cukup data historis
    if len(actual_values) == len(predicted_values):
        mape_score = mape(TimeSeries.from_values(actual_values), TimeSeries.from_values(predicted_values))
        print(f"MAPE for {country} (Prophet): {mape_score:.2f}%")
    else:
        print(f"Skipping MAPE calculation for {country} (Prophet) due to insufficient historical data.")
```

###ARIMA###

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
```

```python
# Buat dictionary untuk menyimpan model ARIMA untuk setiap negara
arima_models = {}
```

```python
# Loop melalui setiap negara dan buat model ARIMA
for country in countries:
    df_country = df[df['Entity'] == country].copy()
    df_country.set_index('Year', inplace=True)

    # Tentukan order (p, d, q) untuk model ARIMA.
    # Ini mungkin memerlukan beberapa eksperimen untuk menemukan order terbaik untuk setiap negara.
    # Contoh di bawah ini menggunakan (5, 1, 0) sebagai titik awal.
    order = (5, 1, 0)

    model = ARIMA(df_country['Primary energy consumption per capita (kWh/person)'], order=order)
    model_fit = model.fit()
    arima_models[country] = model_fit
```

```python
# Prediksi
# Tentukan horizon prediksi (misalnya, 5 tahun ke depan)
horizon = 5
```

```python
# Buat dictionary untuk menyimpan prediksi ARIMA untuk setiap negara
arima_predictions = {}
```

```python
# Loop melalui setiap negara dan buat prediksi
for country, model_fit in arima_models.items():
    # Prediksi untuk horizon yang ditentukan
    forecast = model_fit.forecast(steps=horizon)

    # Simpan prediksi dalam dictionary
    arima_predictions[country] = forecast
```

```python
# Visualisasi
plt.figure(figsize=(12, 6))

for country, prediction in arima_predictions.items():
    historical_series = series_dict[country]

    # Plot data historis
    plt.plot(historical_series.time_index, historical_series.values().flatten(), label=country)

    # Buat indeks waktu untuk prediksi, mulai dari setelah data historis berakhir
    last_historical_time = historical_series.time_index[-1]
    prediction_time_index = pd.date_range(start=last_historical_time + pd.DateOffset(years=1), periods=horizon, freq='Y')

    # Plot prediksi ARIMA
    plt.plot(prediction_time_index, prediction, linestyle='--', color=plt.gca().lines[-1].get_color(), label=f"{country} ARIMA Prediction")

plt.title('Konsumsi Energi Primer per Kapita (ARIMA)')
plt.xlabel('Tahun')
plt.ylabel('kWh/person')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Evaluasi
# Hitung RMSE untuk setiap negara (ARIMA)
for country, prediction in arima_predictions.items():
    historical_series = series_dict[country]
    actual_values = historical_series.values().flatten()[-horizon:]  # Ambil nilai aktual untuk horizon prediksi

    # Hitung RMSE hanya jika ada cukup data historis
    if len(actual_values) == len(prediction):
        rmse = np.sqrt(mean_squared_error(actual_values, prediction))
        print(f"RMSE for {country} (ARIMA): {rmse:.2f}")
    else:
        print(f"Skipping RMSE calculation for {country} (ARIMA) due to insufficient historical data.")
```

```python
# Hitung MAPE untuk setiap negara (ARIMA)
for country, prediction in arima_predictions.items():
    historical_series = series_dict[country]
    actual_values = historical_series.values().flatten()[-horizon:]  # Ambil nilai aktual untuk horizon prediksi

    # Hitung MAPE hanya jika ada cukup data historis
    if len(actual_values) == len(prediction):
        mape_score = mape(TimeSeries.from_values(actual_values), TimeSeries.from_values(prediction))
        print(f"MAPE for {country} (ARIMA): {mape_score:.2f}%")
    else:
        print(f"Skipping MAPE calculation for {country} (ARIMA) due to insufficient historical data.")
```
